```python
def percentage_empty_columns(df: pd.DataFrame, threshold: float) -> list: 
    bad_columns = []
    for name_column, values_column in df.iteritems():
        if df[name_column].isnull().sum()/df.shape[0] >= threshold:
            bad_columns.append(name_column)
    return bad_columns
```

**Колонки с пропусками свыше порога**

```python
threshold = 0.4
drop_columns = percentage_empty_columns(df, threshold)
drop_columns
```

```python
def pie_chart(df:pd.DataFrame, column_name_val: str, column_name, title: str) -> None:
    fig = px.pie(df, values=column_name, names=column_name, title=title)
    fig.show()
```

```python
pie_chart(df, 'Status', 'Stages', 'Launch statuses depending on the number of stages')
```

```python
def scatter_matrix(df: pd.DataFrame, columns: list, color: str, title:str) -> None:
    fig = px.scatter_matrix(df, dimensions=columns, color=color, title=title)
    fig.update_traces(
        showupperhalf=False,
    )
    fig.update_layout(
        height=900,
        
    )
    fig.show()
```

```python
scatter_columns = ["Liftoff Thrust", "Stages", "Payload to LEO", "Strap-ons", "Payload to GTO"]
scatter_matrix(df=df, columns=scatter_columns, color="Status", title="")
```
